```
import FinanceDataReader as fdr

import pandas as pd
import numpy as np

import time
from datetime import datetime

from google.colab import drive
drive.mount('/gdrive', force_remount=True)
```

```
# 전처리 : 입력 변수(재무제표) 불러오기

s_t = time.time()

input_us_fs = pd.read_excel('/gdrive/My Drive/Input_US_FS.xlsx')

# Index 및 Column 값 처리
input_us_fs.index = input_us_fs[input_us_fs.columns[0]].tolist()
input_us_fs = input_us_fs.drop(input_us_fs.columns[0], axis=1)
print(time.time()-s_t)

input_us_fs
```

```
# 전처리 : 출력 변수 (분기 수익률) 처리
# S&P 500 목록 불러오기

snp500 = fdr.StockListing('SP500')
missing = ['OTIS', 'BRKB', 'BFB'] # 누락 종목

snp500_list = list(snp500['Symbol'])

for i in range(len(missing)):
  snp500_list.remove(missing[i])

snp500_list[:10]
```

```
# 종가를 위한 Column 생성
input_us_fs['Close'] = [np.nan]*input_us_fs.shape[0]
```

```
# 종목별 DataFrame 나누기
input_stock = []

j = 0
start = 0
end = 0
for i in range(len(input_us_fs)):
  if snp500_list[j] != input_us_fs.iloc[i].name[:-6]:
    end = i
    input_stock.append(input_us_fs.iloc[start:end])

    start = end
    j += 1

input_stock[0]
```

```
# YYMMDD -> YYYY-MM-DD
def Convert_Date(date): 
  return '20' + date[:2] + '-' + date[2:4] + '-' + date[4:]

# 종가를 얻기 위한 DataFrame 읽어오기
s_t= time.time()
fdr_stock_list = []
for i in range(len(input_stock)):
  start_date = Convert_Date(input_stock[i].iloc[0].name[-6:])
  end_date = Convert_Date(input_stock[i].iloc[-1].name[-6:])
  fdr_stock_list.append(fdr.DataReader(snp500_list[i], start_date, end_date))

print(time.time() - s_t)
fdr_stock_list[0]
```

```
def Next_Month(start):
  month = start[-2:]
  year = start[:4]

  if int(month) == 12:
    year = str(int(year)+1)
    month = '01'
  elif int(month)+1 < 10:
    month = '0' + str(int(month)+1)
  else :
    month = str(int(month)+1)

  return year + '-' + month

# 각 종목연월일에 해당되는 주가의 종가를 가져온다
close_list = []
j = 0 
for i in range(len(input_us_fs)):
  stock_date = input_us_fs.iloc[i].name
  ticker = stock_date[:-6]
  date = Convert_Date(stock_date[-6:])

  try:
    close_list.append(fdr_stock_list[j].loc[date]['Close'])
  except: # 해당 날짜의 주가가 없는 경우
    start = date[:-3]
    end = Next_Month(start)
    temp_df = fdr.DataReader(ticker, start, end)
    try:
      close_list.append(temp_df.iloc[-1]['Close']) # 해당 연월의 마지막 날짜로 추가
    except:
      close_list.append(np.nan)

  # fdr DataFrame의 끝에 도착
  try:
    if ticker != input_us_fs.iloc[i+1].name[:-6]:
      j += 1
      print(j)
  except:
    break

input_us_fs['Close'] = close_list
input_us_fs['Close']
```
